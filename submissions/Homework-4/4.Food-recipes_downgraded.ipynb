```python
import time

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr

import sklearn.cluster as cluster

from sklearn.decomposition import TruncatedSVD

from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
food = open('scirep-cuisines-detail/allr_recipes.txt').read().split('\n')

#Creates initial dictionary for dataframe
foods = {'Total':{}} 


#Parses the text file by line
#Counts the number of times an ingredient is in a recipe by country 
for line in food:
    l = line.split() #
    ctry = l[0] #Grabs the country 
    ingr = l[1:] #Gras the list of ingredients
    
    #Counts total number of recipes per country, used to normalize later
    if (ctry not in foods['Total'].keys()):
        foods['Total'][ctry] = 1
    else:
        foods['Total'][ctry] +=1
 
    for ingrs in ingr:
        #If ingredient is not already in food dict
        if (ingrs not in foods.keys()): 
            foods[ingrs] = {}
        #If country is not in the food dictionary
        if (ctry not in foods[ingrs].keys()):
            foods[ingrs][ctry] = 1 

        else: 
            foods[ingrs][ctry] += 1



#Creates dataframe and fills NaN values with 0 
ingredients_df= pd.DataFrame(foods).fillna(value=0) 

#Normalizes the data so that each ingredient is a percentage
for column in ingredients_df.columns:
    if column is not 'Total':
        ingredients_df[column] = (ingredients_df[column]/ingredients_df['Total'])

ingredients_df
```

```python
#Functions taken from Lecture 9 

#Drops the 'total' column - it's not needed to calculate the linkages
drop_total = ingredients_df.drop('Total', axis=1)

#Calculates the distance matrix 
distanceMatrix = pdist(drop_total,'euclidean')

#Calculates Z 
Z = hr.linkage(drop_total, method='average', metric='euclidean')

fig = plt.figure(figsize=(15,15))
T = hr.dendrogram(Z,color_threshold=1.25, leaf_font_size=10, labels=drop_total.index)
fig.show()
```

Originally, I had the distance calculated using "ward" since it would be less susceptible to noise and outliers. However, this resulted in strange arbitrary clusterings. I decided that using averages to calculate distances would make the most sense since the data consisted of percentages of ingredients used. 

From the above dendogram, we can see clusters as expected such as European countries being clustered together and Asian cultures being clustered togther, though there are also some interesting clusters. for instance, South-African recipes being closely clustered with Belgium. One would expect Belgium to be more closely clustered to Switzerland than South-African recipes. The clustering of Belgium and South-African recipes could also be due to colonization of the latter. Another interesting clustering would be North-African recipes and Turkey. 

We explore these clusterings in more detail below.


```python
#Rotates the original dataframes for easier access 
rotate = ingredients_df.transpose()

print "Number of South-African recipes: ", rotate['South-African']['Total']
print "Number of Belgian recipes: ", rotate['Belgium']['Total']
print "Number of Swiss recipes: ", rotate['Switzerland']['Total']
print


print "Number of North-African recipes: ", rotate['North-African']['Total']
print "Number of Turkish recipes: ", rotate['Turkey']['Total']
print
```

Above, I try to isolate the number of recipes per country to see if the reason they were clustered together was due to the number of recipes each country had. This may be the case with South-African and Belgian recipes since they are closer in total number of recipes to each other than Swiss recipes. However, this doesn't seem to be the case for North-African and Turkish recipes. 

Below, I isolate the most used ingredient for each country. 


```python
#Drops "Total" column to find ingredient used most
drop_t = rotate.drop('Total')

print "Most used ingredient in South Africa: ", drop_t['South-African'].idxmax()
print "Most used ingredient in Belgium: ", drop_t['Belgium'].idxmax()
print "Most used ingredient in Switzerland: ", drop_t['Switzerland'].idxmax()

print

print "Most used ingredient in North Africa: ", drop_t['North-African'].idxmax()
print "Most used ingredient in Turkey: ", drop_t['Turkey'].idxmax()
```

For the case of South Africa, Belgium, and Switzerland, finding the most used ingredient doesn't really indicate how they were clustered together (though previously, it was suggested that the clustering was perhaps due to total number of recipes found in the given data).

Though, for North Africa and Turkey, while they don't share the most used ingredients in their recipes, onions and garlic are quite similar (and are usually used together) which may suggest further similarities in their recipes.
